```python
lam=1064e-9
f=36.6e-3
M2=1.3
MFD= 5.4e-6
w0=MFD/2 #=w0
l = np.linspace(-1e-3, 1e-3, 1001) #delta bei paul
s=f +l

lam_eff=lam**M2
zR=np.pi*w0/lam_eff
q1=s+1j*zR

q2= 1/(1/q1-1/f)

zR_neu=np.imag(q2)

w0_new=np.sqrt(zR_neu*lam_eff/np.pi)

theta      = lam_eff / (np.pi * w0_new)    #halbwinkel
theta_mrad = theta * 1e3                  

def w(distance):
    w=w0*np.sqrt((1+(distance/zR)**2))
    return w


fig, ax=plt.subplots(5,1)
ax[0].plot(l*1e3, w(s),label='w at lens')

plt.show()

```
